```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
X = np.array([1, 2, 3, 4, 5])  # Feature (independent variable)
y = np.array([1.2, 1.8, 2.6, 3.2, 3.8])  # Target (dependent variable)

# Number of data points
n = len(X)

# Calculate the required sums
sum_x = np.sum(X)
sum_y = np.sum(y)
sum_xy = np.sum(X * y)
sum_x2 = np.sum(X ** 2)

# Calculate the slope (m) and intercept (b)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
b = (sum_y - m * sum_x) / n

# Make predictions
y_pred = m * X + b

# Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression (Using Formula)')
plt.xlabel('X (Feature)')
plt.ylabel('y (Target)')
plt.legend()
plt.show()

# Print the results
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
X = np.array([1, 2, 3, 4, 5])  # Feature (independent variable)
y = np.array([1.2, 1.8, 2.6, 3.2, 3.8])  # Target (dependent variable)

# Reshape X to include bias term (column of ones)
X_matrix = np.vstack((np.ones(len(X)), X)).T  # Shape (n, 2)

# Compute coefficients using the Normal Equation: θ = (X^T X)^(-1) X^T y
theta = np.linalg.inv(X_matrix.T @ X_matrix) @ X_matrix.T @ y

# Extract intercept (b) and slope (m)
b, m = theta

# Make predictions
y_pred = X_matrix @ theta

# Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression (Using Matrix Method)')
plt.xlabel('X (Feature)')
plt.ylabel('y (Target)')
plt.legend()
plt.show()

# Print the results
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
```
